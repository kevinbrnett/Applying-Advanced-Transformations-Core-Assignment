```python
df_powers = pd.read_csv("C:\\Users\\hamma\\Documents\\DataEnrichment\\Applying-Advanced-Transformations-Core-Assignment\\superhero_powers - superhero_powers.csv")
df_powers.info()
df_powers.head()
```

```python
# Separate Hero and Publisher
df_info['Hero|Publisher'].str.split('|',expand=True)
```

```python
# Save two new columns to dataframe
df_info[['Hero Names','Publisher']] = df_info['Hero|Publisher'].str.split('|',expand=True)
df_info.head(2)
```

```python
# Drop 'Hero|Publisher' column
df_info = df_info.drop(columns='Hero|Publisher')
df_info.head(2)
```

```python
# Sample of 'Measurements' column
measure = df_info.loc[0,'Measurements']
print(type(measure))
measure
```

```python
# Convert srting dictionary to dictionary
# Replace quotations
measure = measure.replace("'", '"')
measure
```

```python
# Use json.load
fixed_measure = json.loads(measure)
print(type(fixed_measure))
fixed_measure
```

```python
# Apply to whole column
df_info['Measurements'] = df_info['Measurements'].str.replace("'", '"')
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
df_info['Measurements'].head()
```

```python
# Unpack dictionary to create Height and Weight column
hw = df_info['Measurements'].apply(pd.Series)
hw
```

```python
# Concatenate hw with dataframe
df_info = pd.concat((df_info, hw), axis=1)
df_info.head()
```

```python
# Drop 'Measurements' column
df_info = df_info.drop(columns='Measurements')
df_info.head()
```

```python
# Create dictionary with hero name as key and powers as values
powers_dict = df_powers.set_index('hero_names').to_dict()['Powers']
powers_dict
```

```python
# Map powers_dict to df_info to create 'Powers' column
df_info['Powers'] = df_info['Hero Names'].map(powers_dict)
df_info.head()
```

```python
# Split 'Powers' column strings on the ',' and explode to prepare for
# one hot encoding
df_info['Powers_Split'] = df_info['Powers'].str.split(',')
exploded = df_info.explode('Powers_Split')
```

```python
# Variable for columns to make
cols_to_make = exploded['Powers_Split'].dropna().unique()
cols_to_make
```

```python
# For loop to create columns
for col in cols_to_make:
    df_info[col] = df_info['Powers'].str.contains(col)
pd.set_option('display.max_columns', None)
df_info.head()
```

```python
# Drop unecessary columns
df_info = df_info.drop(columns=['Powers','Powers_Split'])
df_info.head()
```

```python
# Remove units from 'Height' and 'Weight' columns and put in column title
df_info = df_info.rename(columns={'Height':'Height_cm','Weight':'Weight_kg'})
df_info.head()
```

```python
df_info['Height_cm'] = df_info['Height_cm'].str.replace(' cm', '')
df_info.head()
```

```python
df_info['Weight_kg'] = df_info['Weight_kg'].str.replace(' kg', '')
df_info.head()
```

```python
# Change datatype to float
df_info['Height_cm'] = df_info['Height_cm'].astype('float')
df_info['Weight_kg'] = df_info['Weight_kg'].astype('float')
```

```python
# Average weight of heroes with and without super speed
df_info.groupby('Super Speed')['Weight_kg'].mean()
```

```python
# Average height of heroes by publisher
df_info.groupby('Publisher')['Height_cm'].mean()
```
