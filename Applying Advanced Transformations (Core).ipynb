```python
# Separate Hero and Publisher
df_info['Hero|Publisher'].str.split('|',expand=True)
```

```python
# Save two new columns to dataframe
df_info[['Hero Names','Publisher']] = df_info['Hero|Publisher'].str.split('|',expand=True)
df_info.head(2)
```

```python
# Drop 'Hero|Publisher' column
df_info = df_info.drop(columns='Hero|Publisher')
df_info.head(2)
```

```python
# Sample of 'Measurements' column
measure = df_info.loc[0,'Measurements']
print(type(measure))
measure
```

```python
# Convert srting dictionary to dictionary
# Replace quotations
measure = measure.replace("'", '"')
measure
```

```python
# Use json.load
fixed_measure = json.loads(measure)
print(type(fixed_measure))
fixed_measure
```

```python
# Apply to whole column
df_info['Measurements'] = df_info['Measurements'].str.replace("'", '"')
df_info['Measurements'] = df_info['Measurements'].apply(json.loads)
df_info['Measurements'].head()
```

```python
# Unpack dictionary to create Height and Weight column
hw = df_info['Measurements'].apply(pd.Series)
hw
```

```python
# Concatenate hw with dataframe
df_info = pd.concat((df_info, hw), axis=1)
df_info.head()
```

```python
# Drop 'Measurements' column
df_info = df_info.drop(columns='Measurements')
df_info.head()
```

```python
# Create dictionary with hero name as key and powers as values
powers_dict = df_powers.set_index('hero_names').to_dict()['Powers']
```

```python
# Map powers_dict to df_info to create 'Powers' column
df_info['Powers'] = df_info['Hero Names'].map(powers_dict)
df_info.head()
```
